```python
import sys
from typing import List

def main() -> None:
    data: List[str] = sys.stdin.read().strip().split()
    if not data:
        return

    it = iter(data)

    n: int = int(next(it))
    m: int = int(next(it))

    adj: List[List[int]] = [[] for _ in range(n)]
    for _ in range(m):
        u: int = int(next(it))
        v: int = int(next(it))
        adj[u].append(v)
        adj[v].append(u)

    visited: List[bool] = [False] * n

    components: List[List[int]] = []

    for start in range(n):
        if not visited[start]:
            stack: List[int] = [start]
            visited[start] = True
            current_component: List[int] = []

            while stack:
                v: int = stack.pop()
                current_component.append(v)

                for nei in adj[v]:
                    if not visited[nei]:
                        visited[nei] = True
                        stack.append(nei)

            components.append(current_component)

    out_lines: List[str] = []
    out_lines.append(str(len(components)))

    for comp in components:
        out_lines.append(str(len(comp)))
        out_lines.append(" ".join(str(v) for v in comp))

    sys.stdout.write("\n".join(out_lines))


if __name__ == "__main__":
    main()
```

```python
input_01 = [
    (3, 2), # N, M
    (0, 1), # v_1, u_1
    (1, 2), # v_2, u_2
]
```

```python
from copy import deepcopy

N, M = input_01[0][:]
print(f"Vertices (N): {N} | Edges (M): {M}")

G = [[] for _ in range(N)]

for i in range(0, M):
    u, v = input_01[i + 1][:]
    G[u].append(v)
    G[v].append(u)

G_ = deepcopy(G)

for i in range(len(G_)):
    print(f"{i}: {G[i]}")
```

```python
components = []
visited = [False] * N

for start in range(N):

    stack = [start]
    visited[start] = True
    current_component = []

    while stack:
        v = stack.pop()
        current_component.append(v)

        for neighbours in G_[v]:
            if not visited[neighbours]:
                visited[neighbours] = True
                stack.append(neighbours)

    components.append(current_component)
        

```
