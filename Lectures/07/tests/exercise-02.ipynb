```python
import collections
import sys

def solve():
    try:
        line = sys.stdin.readline()
        if not line:
            return
        N, M = map(int, line.split())
    except EOFError:
        return
    except ValueError:
        return

    park_map = []
    start_pos = None
    
    for r in range(N):
        try:
            row = sys.stdin.readline().strip()
            if len(row) != M:
                continue 

            park_map.append(row)

            if 'E' in row:
                c = row.index('E')
                start_pos = (r, c)
        except EOFError:
            break

    if start_pos is None:
        print(-1)
        return

    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    queue = collections.deque([start_pos])

    dist = [[-1] * M for _ in range(N)]
    dist[start_pos[0]][start_pos[1]] = 0

    while queue:
        r, c = queue.popleft()
        current_distance = dist[r][c]

        if park_map[r][c] == 'X':
            print(current_distance)
            return

        for dr, dc in directions:
            nr, nc = r + dr, c + dc

            is_in_bounds = (0 <= nr < N) and (0 <= nc < M)
            
            if is_in_bounds:
                cell_content = park_map[nr][nc]
                is_barrier = cell_content == '#'
                is_unvisited = dist[nr][nc] == -1
                
                if not is_barrier and is_unvisited:
                    dist[nr][nc] = current_distance + 1
                    queue.append((nr, nc))

    print(-1)

if __name__ == "__main__":
    solve()
```
