```python
import sys

content_input = read_data()
content_output = read_data("./output.txt")

N, M = map(int, content_input[0].split())
scores = list(map(int, content_input[1].split()))
lr = [tuple(map(int, line.split())) for line in content_input[2:M+2]]

print(lr)
```

```python
content_output = [int(line.strip()) for line in content_output]
content_output
```

```python
results = [0] * M

for i, interval in enumerate(lr):

    min_val, max_val = min(interval), max(interval)

    for score in scores:
        if min_val <= score <= max_val:
            results[i] += 1
    print(results[i])
 
results
```

```python
def within_interval(value: int, min_value: int, max_value: int) -> bool:
    return min_value <= value <= max_value

def filter_scores_in_interval(scores: list[int], interval: tuple[int, int]) -> list[int]:
    return list(filter(lambda score: within_interval(score, min(interval), max(interval)), scores))

for i, interval in enumerate(lr):
    filtered_scores = filter_scores_in_interval(scores, interval)
    results[i] = len(filtered_scores)
    print(results[i])
    
```
