```python
 
quantum_a = 7 ** 5
quantum_m = 2 ** 31 - 1

def counting_sort(x: list) -> list:
    count = {}
    for value in x:
        if value in count:
            count[value] += 1
        else:
            count[value] = 1

    sorted_x = []
    
    for key in sorted(count.keys()):
        sorted_x.extend([key] * count[key])
    return sorted_x


def analyze_trimpazation(n: int, m: int, q0: int) -> int:
    '''
    This function generates data with given parameters
    and calculates desired Y value.

    You need to modify it to make it execute faster.
    You can check your progress using estimate_execution_time flag
    at the top of the file.
    '''
    m_div2 = m // 2
    q = q0

    counts = [0] * m 

    # Generating data and counting occurrences:
    for _ in range(n):
        
        x_i = q % m - m_div2
        counts[x_i + m_div2] += 1
        q = (q * quantum_a) % quantum_m

    # Weighted sum from counts:
    response = 0
    position = 1

    for index, count in enumerate(counts):
        if count == 0:
            continue

        value = index - m_div2
        response += value * (count * position + count * (count - 1) // 2 )
        position += count

    return response

if __name__ == '__main__':
    # N, M, q0 = map(int, input().split())
    executions_times = []
    N, M, q0 = prepare_data()
    
    print(N, M, q0)
    for i in range(5):
        start = time()
        print(analyze_trimpazation(N, M, q0))
        end = time()
        executions_times.append(end - start)

    if estimate_execution_time:
        for exec_time in executions_times:
            print(f'Execution time = {exec_time:.8f} seconds')
        
```

```python
from bisect import bisect_left, bisect_right

raw_x = [0, 2, -1,  1, 3, 1, -4, 1, 0, 0]
x = []

for value in raw_x:
    
    print(f'Inserting value: {value}')

    if len(x) == 0:
        x.append(value)
        print("\t  - First element added.")

    else:

        min_val, max_val = x[0], x[-1]
        print(f'\tCurrent min: {min_val}, max: {max_val}')

        if value < min_val:
            print(f'\t  - Inserting {value} at the beginning.')
            x.insert(0, value)

        elif value >= max_val:
            x.append(value)
            print(f'\t  - Appending {value} at the end.')

        else:
            position = bisect_right(x, value)
            x.insert(position, value)
            print(f'\t  - Inserting {value} at position {position}.')

    print('-' * 100)
    print(f'Current x: {x}')
    print('-' * 100)
    print()
x
```
