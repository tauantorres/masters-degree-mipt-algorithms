```python
from typing import Tuple, List


INPUT_DATA: str = "./input.txt"

def read_data(FILE_PATH: str = INPUT_DATA) -> list[str]:
    with open(FILE_PATH, 'r') as file:
        lines = file.readlines()
    return lines

def prepare_data() -> Tuple[int, int, List[int]]:

    content_input = read_data()

    # Exercise Rules:
    N, M = map(int, content_input[0].split())
    coordinates = list(map(int, content_input[1].split()))
    
    return N, M, coordinates

# N: number of sittings
# M: number of students
# positions: Coordinates of sittings

def can_place_enough_students(positions: List[int], distance: int, min_num_students: int) -> bool:

    count, current_possition = 1, positions[0]

    for position in positions[1:]:
        if (position - current_possition) >= distance:

            count += 1
            current_possition = position

            if count >= min_num_students:
                return True

    return count >= min_num_students

def maximize_min_distance(positions: List[int], min_num_students: int) -> int:

    answer = 0
    lowest_value = 0
    max_distance = positions[-1] - positions[0]

    while lowest_value <= max_distance:

        middle = (lowest_value + max_distance) // 2

        if can_place_enough_students(positions=positions, distance=middle, min_num_students=min_num_students):
            answer, lowest_value = middle, middle + 1

        else:
            max_distance = middle - 1

    return answer

def main() -> None:
    N, M, positions = prepare_data()
    positions = sorted(positions)
    answer = maximize_min_distance(positions=positions, min_num_students=M)
    print(answer)

```

```python
def binary_search_exists(X: List[int], key: int) -> bool:
    left, right = 0, len(X)
    print(f"Array: {X}")
    print(f"Initial Left: {left}, Right: {right}, Key: {key}")
    while (right - left) > 1:
        print(f"-" *50)

        middle = (left + right) // 2
        print(f"Left: {left}, Right: {right}, Middle: {middle}, X[Middle]: {X[middle]}")

        if X[middle] <= key:
            left = middle

            print(f"\t - [{X[middle]} <= {key} ? {X[middle] <= key}] Updated Left to {left}")

        else:
            right = middle
            print(f"\t - [{X[middle]} <= {key} ? {X[middle] <= key}] Updated Right to {right}")

    print(f"-" *50)
    print(f"\nFinal Left: {left}, X[Left]: {X[left]}, Key: {key}")

    return X[left] == key

X = [1, 2, 3, 4, 5, 6, 7]
key = 3

print(binary_search_exists(X, key))
```
