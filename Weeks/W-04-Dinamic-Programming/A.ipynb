```python
from typing import List
from bisect import bisect_left

def LIS(N: int, X: List[int]) -> int:

    d = [float('inf')] * (N + 1)
    d[0] = -float('inf')

    for i in range(N):
        l = bisect_left(d, X[i])

        if d[l - 1] < X[i] < d[l]:
            d[l] = X[i]

    answer = max(
        i for i in range(N + 1) 
        if d[i] < float('inf')
    )

    return answer, d

if __name__ == '__main__':
    # N = int(input())
    # X = list(map(int, input().split()))

    N, X = prepare_data()
    print(LIS(N, X))

```
