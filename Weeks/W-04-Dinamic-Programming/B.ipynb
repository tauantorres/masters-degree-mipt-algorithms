```python
from typing import List

def solve_nails(N: int, X: List[int]) -> int:

    X = sorted(X)

    if N == 2:
        return X[1] - X[0]

    d = [0] * N

    d[0] = 10 ** 9 
    d[1] = X[1] - X[0]

    for i in range(2, N):
        d[i] = min(d[i -1], d[i - 2]) + (X[i] - X[i - 1])

    return d[N - 1]

if __name__ == '__main__':
    # N = int(input())
    # X = list(map(int, input().split()))
    N, X = prepare_data()
    print(solve_nails(N, X))
```
