```python
from typing import List

def knapsack(W: int, N: int, w: List[int], c: List[int]) -> int:
    d = [0] * (W + 1)

    for i in range(N):
        weight = w[i]
        cost = c[i]

        for j in range(W, weight - 1, -1):
            d[j] = max(d[j], d[j - weight] + cost)

    return d[W]

if __name__ == '__main__':

    # W, N = tuple(map(int, input().split()))
    # w = list(map(int, input().split()))
    # c = list(map(int, input().split()))

    W, N, w, c = prepare_data()
    print(knapsack(W, N, w, c))

```

```python
from typing import List


def knapsack(Weight: int, Items: int, weights: List[int], costs: List[int]) -> int:

    d = [ [0] * (Weight + 1) for _ in range(Items + 1) ]

    for item in range(1, Items + 1):
        for weight in range(1, Weight + 1):
            if weights[item - 1] > weight:
                d[item][weight] = d[item - 1][weight]
            else:
                d[item][weight] = max(
                    d[item - 1][weight],
                    d[item - 1][weight - weights[item - 1]] + costs[item - 1]
                )

    w_remaining = Weight
    position_bars_added = []

    for item in range(Items, 0, -1):
        if d[item][w_remaining] != d[item - 1][w_remaining]:
            position_bars_added.append(item)
            w_remaining -= weights[item - 1]

    position_bars_added.reverse()
    total_bars_added = len(position_bars_added)

    print(d[Items][Weight]) 
    print(total_bars_added)
    print(*position_bars_added)


if __name__ == '__main__':

    # W, N = tuple(map(int, input().split()))
    # w = list(map(int, input().split()))
    # c = list(map(int, input().split()))

    W, N, w, c = prepare_data()
    knapsack(W, N, w, c)

```
