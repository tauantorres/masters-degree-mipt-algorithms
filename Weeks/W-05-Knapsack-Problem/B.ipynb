```python
from typing import List


def reverse_knapsack(Items: int, costs: List[int]) -> None:

    total_cost = sum(costs)

    if total_cost % 2 != 0:
        print("NO")
    else:
        subarray_sum_target = total_cost // 2

        d = [[False] * (subarray_sum_target + 1) for _ in range(Items + 1)]
        d[0][0] = True

        for item in range(1, Items + 1):
            for cost in range(subarray_sum_target + 1):
                d[item][cost] = d[item - 1][cost]
                if cost >= costs[item - 1]:
                    d[item][cost] = d[item][cost] or d[item - 1][cost - costs[item - 1]]

        if d[Items][subarray_sum_target]:
            print("YES")
        else:
            print("NO")


if __name__ == '__main__':

    Items = int(input())
    costs = list(map(int, input().split()))

    Items, costs = prepare_data()
    reverse_knapsack(Items, costs)
```
