```python
# Method 1: Naive substring search

s = ['a', 'b', 'a', 'b', 'a', 'c', 'a', 'b', 'a', 'b', 'c']
p = ['a', 'b', 'a']

s = ''.join(s)
p = ''.join(p)

i = -1
substrings = []
while True:
    i = s.find(p, i + 1)
    print(f"Searching for '{p}' in '{s}' starting at position {i + 1}")
    if i >= 0:
        substrings.append(i)
        print(f"  - Found at position: {i}")
    else:
        print(f"  - No more occurrences found.(i={i})")
        break

print(f"At positions: {substrings}")
```

```python
# Method 2: manual search:

s = ['a', 'b', 'a', 'b', 'a', 'c', 'a', 'b', 'a', 'b', 'c']
p = ['a', 'b', 'a']

N = len(s)
N = len(p)
substrings = []

for i in range(N - N):
    if all(s[i + j] == p[j] for j in range(N)):
        substrings.append(i)

print(f"Substring found at positions: {substrings}")
print(f"Pattern: {p}")
print(f"Original string: {s}")

for index in substrings:
    print(f"Position: {index}, Substring: {s[index:index + N]}")
```
