```python
from typing import Self, List, Tuple

class BuildingHeap:

    def __init__(self: Self) -> None:
        self.data: List[Tuple[str, int, int]] = []

    def _get_parent_index(self: Self, index: int) -> int:
        return (index - 1) // 2

    def _get_children_indices(self: Self, index: int) -> Tuple[int, int]:
        return 2 * index + 1, 2 * index + 2

    def sift_up(self: Self, index: int) -> None:

        if index == 0:
            return

        parent = self._get_parent_index(index)

        if self.data[parent] > self.data[index]:
            self.data[parent], self.data[index] = self.data[index], self.data[parent]
            self.sift_up(parent)

    def sift_down(self: Self, index: int) -> None:

        N = len(self.data)
        left_node, right_node = self._get_children_indices(index)

        if left_node >= N:
            return
        
        if right_node >= N:
            child_min = left_node
        else:
            child_min = left_node if self.data[left_node] < self.data[right_node] else right_node

        if self.data[child_min] < self.data[index]:
            self.data[index], self.data[child_min] = self.data[child_min], self.data[index]
            self.sift_down(child_min)

    def push(self: Self, item: Tuple[str, int, int]) -> None:
        self.data.append(item)
        self.sift_up(len(self.data) - 1)

    def pop(self: Self) -> Tuple[str, int, int]:
        top, last = self.data[0], self.data.pop()

        if self.data:
            self.data[0] = last
            self.sift_down(0)

        return top

    def __len__(self: Self) -> int:
        return len(self.data)
    
heap = BuildingHeap()
order = 0

for event in Events:

    event = event.split(' ')
    event_type = event[0]

    if event_type == '+':
        costumer_id, priority = int(event[1]), int(event[2])
        heap.push( ( -priority, order, costumer_id ) )
        order += 1
    else:
        _, _, costumer_id = heap.pop()
        print(costumer_id)
```
